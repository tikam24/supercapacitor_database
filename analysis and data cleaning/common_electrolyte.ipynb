```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_excel("merged_file.xlsx") # read excel file
df.head()
```

```python
material = "electrolyte"
```

```python
unique_materials = df[material].unique()
# print(unique_material)
unique_df = pd.DataFrame({f'Unique_{material}': unique_materials})
unique_df.to_excel(f"{material}_materials_list.xlsx", index=False, engine ="openpyxl")
```

```python
material_counts = df[material].value_counts()
print(material_counts)
counts_df = pd.DataFrame({f'Unique_{material}': material_counts.index, 'Counts': material_counts.values})
counts_df.to_excel(f"{material}_count_list.xlsx", index=False)
```
