### Code to delete the rows for some particular columns having empty cells

```python
import pandas as pd
```

```python
df = pd.read_excel("prop2.xlsx")
df.shape
df.head()
```

```python
# Specify the columns in which you want to check for empty values
columns_to_check = ['Specific Capacitance Value','Specific Capacitance Unit', 'Surface Area Value', 'Surface Area Unit',
                    'Energy Density Value','Energy Density Unit', 'Power Density Value','Power Density Unit', 'Voltage Value',
                    'Voltage Unit', 'Current Density Value','Current Density Unit', 'Cycles Value', 'Cycles Unit']
```

```python
# Drop rows with empty values in specified columns
df.dropna(subset=columns_to_check, how='all', inplace=True)
```

```python
# Save to excel file
df.to_excel('properties2.xlsx', index=False)
```
