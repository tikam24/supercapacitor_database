```python
def coma_split(item):
    try:
        if item is not None:
            values = item.split(', ')
            float_value = float(values[0])
            return float_value
        else:
            return None
    except AttributeError:
        return None        
```

```python
def count_num(item):
    try:
        if item is not None:
            values = item.split(', ')
            float_value = len(values)
            return float_value
        else:
            return 0
    except AttributeError:
        return 1       
```

```python
z = df[f"{i}_count"] = df[i].apply(count_num)
```
